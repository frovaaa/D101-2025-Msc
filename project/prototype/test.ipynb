```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt
```

```python

with rasterio.open("nuovaZelanda.tif") as src:
    dem = src.read(1)
plt.imshow(dem, cmap="terrain")
plt.colorbar(label="Elevation (m)")
plt.show()
```

```python
# Let's check the New Zealand DEM size and extent
with rasterio.open("nuovaZelanda.tif") as src:
    print(f"DEM size: {src.width} x {src.height} pixels")
    print(f"Pixel size: {src.res[0]:.2f} x {src.res[1]:.2f} meters")
    print(f"Total area: {src.width * src.res[0] / 1000:.1f} x {src.height * src.res[1] / 1000:.1f} km")
    print(f"Bounds: {src.bounds}")
    print(f"CRS: {src.crs}")
    
    # Calculate how much data this would be
    total_pixels = src.width * src.height
    print(f"\nTotal pixels: {total_pixels:,}")
    print(f"At 50m resolution, this would be: {total_pixels * 2 / (1024**3):.1f} GB for heights.bin")
    
    # Show bounds in a more readable format
    print(f"\nBounding box (approx):")
    print(f"  West:  {src.bounds.left:.1f}")
    print(f"  East:  {src.bounds.right:.1f}")
    print(f"  South: {src.bounds.bottom:.1f}")
    print(f"  North: {src.bounds.top:.1f}")
```
